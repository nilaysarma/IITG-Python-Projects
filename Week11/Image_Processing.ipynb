# Week 11 Lab: Image Processing (Grayscale & Black-White Conversion)
### Objective: Convert images into grayscale and binary (black and white) using OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
```

1. Load a color image using `cv2.imread()` .

```python
image = cv2.imread("image.png")
```

2. Convert the image to grayscale using `cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)` .

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

3. Convert the grayscale image to a binary black-and-white image using thresholding
(`cv2.threshold()`).

```python
threshold_value = 128  # Set the threshold value
max_value = 255  # Maximum pixel value
_, binary_image = cv2.threshold(gray_image, threshold_value, max_value, cv2.THRESH_BINARY)
```

4. Display the original, grayscale, and binary images using `matplotlib.pyplot.imshow()` .

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))

# Show original image
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Show grayscale image
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

# Show binary image
plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap="gray")
plt.title("Binary Image")
plt.axis("off")

# Show the images
plt.tight_layout()
plt.show()
```

5. Save the grayscale and black-and-white images using `cv2.imwrite()` .

```python
cv2.imwrite("grayscale_image.jpg", gray_image)
cv2.imwrite("binary_image.jpg", binary_image)
```
