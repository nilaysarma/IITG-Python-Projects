# Week 11 Lab: Sound Processing with Librosa
### Objective: Work with audio files using Librosa to mix and manipulate sounds.

```python
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np
from IPython.display import Audio
```

1. Load a .wav file using `librosa.load()` .

```python
filename = "lash.wav"
    
audio, sr = librosa.load(filename, sr=None)
```

2. Plot the waveform using `matplotlib.pyplot.plot()` .

```python
plt.figure(figsize=(10, 4))
plt.plot(audio, color='blue')
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Waveform of Audio File")
plt.show()
```

3. Load another .wav file and mix it with the first one by summing the signals.

```python
filename2 = "VideoGameRisingSFX.wav"

audio2, sr2 = librosa.load(filename2, sr=None)
```

```python
min_length = min(len(audio), len(audio2))
audio = audio[:min_length]
audio2 = audio2[:min_length]
```

```python
mixed_audio = audio + audio2

plt.figure(figsize=(10, 4))
plt.plot(mixed_audio, color='purple')
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Waveform of Mixed Audio")
plt.show()
```

4. Normalize the mixed signal and save it back as a .wav file using `soundfile.write()` .

```python
output_filename = "mixed_audio.wav"

# Normalize the mixed audio
mixed_audio = mixed_audio / np.max(np.abs(mixed_audio))
    
# Save the normalized mixed audio as a .wav file
sf.write(output_filename, mixed_audio, sr)
    
# Plot the waveform of the mixed audio
plt.figure(figsize=(10, 4))
plt.plot(mixed_audio, color='purple')
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.title("Waveform of Normalized Mixed Audio")
plt.show()
```

5. Play the mixed audio using an external library like `IPython.display.Audio()` .

```python
Audio(mixed_audio, rate=sr)
```
